```python
import pandas as pd
import numpy as np

people = pd.read_csv('/content/drive/MyDrive/SOC-2026/Assignment-1/Q1/unique_people_dataset_2000.csv')
teachers   = pd.read_csv('/content/drive/MyDrive/SOC-2026/Assignment-1/Q1/all_teacher.csv')

people.head()
```

```python
teachers.head()
```

##Suspect filtering on people

```python
suspects = people[
    (people['age'] >= 30) & (people['age'] <= 45) &
    (people['height'] >= 160) & (people['height'] <= 190) &
    (people['weight'] >= 50) & (people['weight'] <= 100) &
    (people['location'].isin(['city_south']))
]
print('Number of suspects:', suspects.shape[0])
suspects
```

##Suspect filtering on teachers

```python
suspects = teachers[
    (teachers['age'] >= 30) & (teachers['age'] <= 45) &
    (teachers['height'] >= 160) & (teachers['height'] <= 190) &
    (teachers['weight'] >= 50) & (teachers['weight'] <= 100) &
    (teachers['location'].isin(['city_south']))
]
print('Number of suspects:', suspects.shape[0])
suspects
```

##"close friend since long"

```python
suspects = suspects[suspects['Joining year'] <= 2014]
print('After joining-year filter:', suspects.shape[0])
suspects
```
