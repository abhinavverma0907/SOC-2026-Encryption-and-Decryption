## SET-UP

```python
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j',
            'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't',
            'u', 'v', 'w', 'x', 'y', 'z', ' ', '.', ',', ';',
            ':', '/', '@']

char_to_idx = {ch: i for i, ch in enumerate(alphabet)}
N = len(alphabet)

def decrypt(text, shift):
    result = ""
    for ch in text:
        i = char_to_idx[ch]
        new_i = (i - shift) % N
        result += alphabet[new_i]
    return result

def frequency_analysis(text):
    # Count each character
    freq = {}
    for ch in text:
        freq[ch] = freq.get(ch, 0) + 1
    # Sort by frequency (most common first)
    return sorted(freq.items(), key=lambda x: x[1], reverse=True)

print("Setup complete")
```

### Method

### Historical Note
Both messages were encrypted with the same Caesar shift (17). This made Thomas Phelippes' job easier
— he could forge the postscript using the same cipher key.
