## SET-UP

```python
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j',
            'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't',
            'u', 'v', 'w', 'x', 'y', 'z', ' ', '.', ',', ';',
            ':', '/', '@']

N = len(alphabet)
char_to_idx = {ch: i for i, ch in enumerate(alphabet)}

def decrypt(text, shift):
    result = ""
    for ch in text:
        i = char_to_idx[ch]
        result += alphabet[(i - shift) % N]
    return result

print("Setup complete. Alphabet size:", N)
```

```python
def frequency_analysis(ciphertext):
    freq = {}
    for ch in ciphertext:
        freq[ch] = freq.get(ch, 0) + 1

    sorted_freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)

    print("Top 5 most frequent characters:")
    for ch, count in sorted_freq[:5]:
        print(f"  '{ch}' → {count} times")

    most_common = sorted_freq[0][0]
    print(f"\nMost common cipher char: '{most_common}'")

    shift = (char_to_idx[most_common] - char_to_idx[' ']) % N
    print(f"Predicted shift: {shift}")

    return shift
```

## Q2 — Caesar Cipher Decryption

### Method: Frequency Analysis
In English text, **space** is the most common character (~13% of any normal text).
In a Caesar cipher every character shifts by the same fixed amount.
So the most frequent character in the ciphertext must be the encrypted version of space.

Steps:
1. Count frequency of every character in the ciphertext
2. Identify the most common one
3. Compute shift = (index of most common char) - (index of space) mod 33
4. Decrypt using that shift — no brute force needed
